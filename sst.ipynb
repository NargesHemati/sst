```python
import speech_recognition as sr
```

```python
stt = sr.Recognizer()
seda3 = sr.AudioFile('seda3.wav')

with seda3 as source:
    audio = stt.record(source)
    
text = stt.recognize_google(audio, language ='fa-IR')
print (text)
```

```python
stt = sr.Recognizer()
with sr.Microphone() as source:
    stt.pause_threshold = 1
    audio = stt.listen(source)
    
text = stt.recognize_google(audio, language ='fa-IR')
print (text)
```
